```python
s# User to enter the number of days
days = int(input("Enter the number of days: "))
# Calculate number of seconds in the given days
seconds = days * 24 * 60 * 60
# Display the result
print(f"There are {seconds} seconds in {days} days.")
```

```python
# User input a character
character = input("Enter a character: ")
# Check if the input is a upper case or lower case
if character.isupper():
    print(f"{character} is an upper case letter.")
elif character.islower():
    print(f"{character} is a lower case letter.")
else:
    print(f"{character} is not a letter.")
    # Display results
    print(f"{character} is an upper case letter.")
```

```python
# User to enter radius of a circle
radius = float(input("Enter the radius of the circle: "))
#calculate volume of the circle using radius give
volume = (4/3) * 3.14159 * (radius ** 3)
# Display the result
print(f"The volume of the circle with radius {radius} is {volume:.2f} cubic units.")
```

```python
# Enter length and width of a square
length = float(input("Enter the length of the square: "))
width = float(input("Enter the width of the square: "))
# Calculate and perimeter, area of the square
perimeter = 2 * (length + width)
area = length * width
# Display results
print(f"The perimeter of the square is {perimeter:.2f} units.")
print(f"The area of the square is {area:.2f}2 square units.")
```

```python
x = 0
y = 20

# Repeat until y is less than 6
while y >= 6:
    y = y - 4        # subtract 4 from y
    x = x + (2 / y)  # add 2/y to x

print("Final value of x:", x)
```

```python
def main():
    numbers = []   # empty list (array) to store values

    print("Enter 5 numbers:")

    # Loop 5 times to get input
    for i in range(5):
        value = float(input(f"Enter value {i+1}: "))
        numbers.append(value)

    # Calculate average
    average = sum(numbers) / len(numbers)

    # Display result
    print("The numbers you entered are:", numbers)
    print("The average is:", average)

# Run the program
main()
```
