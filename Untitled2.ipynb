# Task

## Descriptive statistics

### Subtask:
Calculate and display descriptive statistics for numerical columns to understand the central tendency, dispersion, and shape of the data.

## Visualize distributions

### Subtask:
Create visualizations (e.g., histograms, box plots) for numerical columns to observe their distributions and identify potential outliers.


**Reasoning**:
Create visualizations (histograms and box plots) for numerical columns to observe their distributions and identify potential outliers as requested by the subtask instructions.



```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = df.select_dtypes(include=np.number).columns

for col in numerical_cols:
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    sns.histplot(data=df, x=col, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

    plt.subplot(1, 2, 2)
    sns.boxplot(data=df, y=col)
    plt.title(f'Box Plot of {col}')
    plt.ylabel(col)

    plt.tight_layout()
    plt.show()
```

## Analyze categorical data

### Subtask:
Analyze the distribution of categorical variables (e.g., Gender) using frequency counts and possibly visualizations (e.g., bar plots).


**Reasoning**:
Select categorical columns and then for each categorical column, calculate frequency counts and create a bar plot to visualize the distribution.



```python
import matplotlib.pyplot as plt
import seaborn as sns

categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    print(f"Frequency counts for column: {col}")
    display(df[col].value_counts())

    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=col)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

### Subtask:
Investigate relationships between different variables using scatter plots, correlation matrices, or other relevant visualizations to identify potential patterns and insights.


**Reasoning**:
Create scatter plots to visualize the relationships between 'Annual Income (k$)' and 'Spending Score (1-100)', and between 'Age' and 'Annual Income (k$)', and 'Age' and 'Spending Score (1-100)', differentiating by gender in the first plot. Also, calculate and visualize the correlation matrix for the numerical columns.



```python
# Scatter plot of 'Annual Income (k$)' vs 'Spending Score (1-100)' by Gender
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Gender')
plt.title('Annual Income vs Spending Score by Gender')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()

# Scatter plot of 'Age' vs 'Annual Income (k$)'
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Annual Income (k$)')
plt.title('Age vs Annual Income')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.show()

# Scatter plot of 'Age' vs 'Spending Score (1-100)'
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Spending Score (1-100)')
plt.title('Age vs Spending Score')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.show()

# Calculate the correlation matrix
numerical_cols = df.select_dtypes(include=np.number)
correlation_matrix = numerical_cols.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

# Task
Perform k-means clustering on the dataframe `df` using:
a. 1 suitable feature.
b. 2 suitable features.
Visualize the results for both cases.

## Select features for 1-feature clustering

### Subtask:
Choose one suitable numerical feature for clustering.


**Reasoning**:
Examine the numerical columns and select one suitable feature for 1D clustering based on the previous EDA. The 'Spending Score (1-100)' column shows a relatively spread-out distribution and potential clusters, making it suitable for 1D clustering.



```python
feature_1d = 'Spending Score (1-100)'
print(f"Selected feature for 1D clustering: {feature_1d}")
```

## Apply k-means (1 feature)

### Subtask:
Apply the K-Means algorithm to the selected feature.


**Reasoning**:
Apply the K-Means algorithm to the selected feature as requested by the subtask instructions.



```python
from sklearn.cluster import KMeans

feature_1d_data = df[[feature_1d]].values

# Determine an appropriate number of clusters (e.g., based on visual inspection of the distribution)
# Let's start with 4 clusters for the Spending Score based on the histogram
n_clusters_1d = 4

kmeans_1d = KMeans(n_clusters=n_clusters_1d, random_state=42, n_init=10)
df['cluster_1d'] = kmeans_1d.fit_predict(feature_1d_data)

display(df.head())
```

## Visualize 1-feature clusters

### Subtask:
Visualize the clusters created using the single feature.


**Reasoning**:
Create a scatter plot of the selected feature ('Spending Score (1-100)') against the CustomerID, color the data points by cluster, add labels and a title, and display the plot as requested by the subtask instructions.



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='CustomerID', y=feature_1d, hue='cluster_1d', palette='viridis', legend='full')
plt.title(f'K-Means Clustering on {feature_1d}')
plt.xlabel('Customer ID')
plt.ylabel(feature_1d)
plt.show()
```

## Select features for 2-feature clustering

### Subtask:
Choose two suitable numerical features for clustering.


**Reasoning**:
Based on the scatter plots in the EDA, 'Annual Income (k$)' and 'Spending Score (1-100)' showed potential clustering patterns, making them suitable features for 2-feature clustering to identify customer segments.



```python
feature_2d = ['Annual Income (k$)', 'Spending Score (1-100)']
print(f"Selected features for 2D clustering: {feature_2d}")
```

## Apply k-means (2 features)

### Subtask:
Apply the K-Means algorithm to the two selected features.


**Reasoning**:
Select the data for the two chosen features, import KMeans, determine the number of clusters based on the scatter plot from EDA, instantiate and fit the KMeans model, and add the cluster labels to the dataframe as requested by the subtask instructions.



```python
from sklearn.cluster import KMeans

features_2d_data = df[feature_2d].values

# Based on the scatter plot of Annual Income vs Spending Score, 5 clusters seem appropriate.
n_clusters_2d = 5

kmeans_2d = KMeans(n_clusters=n_clusters_2d, random_state=42, n_init=10)
df['cluster_2d'] = kmeans_2d.fit_predict(features_2d_data)

display(df.head())
```

## Visualize 2-feature clusters

### Subtask:
Visualize the clusters created using the two features.


**Reasoning**:
Create a scatter plot to visualize the clusters using 'Annual Income (k$)' and 'Spending Score (1-100)', colored by 'cluster_2d', as requested by the subtask instructions.



```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='cluster_2d', palette='viridis', legend='full')
plt.title('K-Means Clustering on Annual Income and Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

## Summary:

### Data Analysis Key Findings

*   For 1-feature clustering, 'Spending Score (1-100)' was selected as a suitable feature. K-Means was applied with 4 clusters, and the resulting clusters were visualized against Customer ID.
*   For 2-feature clustering, 'Annual Income (k\$)' and 'Spending Score (1-100)' were selected as suitable features. K-Means was applied with 5 clusters, and the resulting clusters were visualized.

### Insights or Next Steps

*   The 2-feature clustering on 'Annual Income (k\$)' and 'Spending Score (1-100)' appears to identify more distinct customer segments than the 1-feature clustering on 'Spending Score (1-100)' alone, suggesting that considering both features provides better separation.
*   Further analysis could involve characterizing the customer segments identified in the 2-feature clustering based on other demographic information (e.g., Age, Gender) to gain deeper insights into their profiles.
